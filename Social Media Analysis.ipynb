```python

Berry['Date'] = pd.to_datetime(Berry['Date'])
Berry['Hour'] = Berry['Date'].dt.hour

bins = [0, 2, 4, 6, 8, 10, 13, 19, 21, 24]

Berry['Time Slot'] = pd.cut(Berry['Hour'], bins=bins[:-1], labels=['22:00-00:00', '00:00-02:00', '02:00-04:00', '04:00-06:00',
                                                              '06:00-08:00', '11:00-13:00', '19:00-21:00', '21:00-22:00'], right=False)

result_df = Berry.groupby('Time Slot')[['Insights - Reach', 'Insights - Impressions']].mean().round(2)

pd.DataFrame(result_df)
```
